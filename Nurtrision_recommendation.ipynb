# Manually Calculating the Metrics

```python
# Function to get user inputs
def get_user_inputs():
    age = int(input("Enter your age: "))
    gender = input("Enter your gender (male/female): ").strip().lower()
    height_cm = float(input("Enter your height in centimeters: "))
    weight_kg = float(input("Enter your weight in kilograms: "))
    weight_goal = input("Enter your weight goal (loss/gain): ").strip().lower()
    activity_level = input("Enter your activity level (sedentary/light/moderate/active/very active): ").strip().lower()
    time_frame_months = int(input("Enter the time frame in months to achieve your goal: "))
    return age, gender, height_cm, weight_kg, weight_goal, activity_level, time_frame_months

# Get user inputs
age, gender, height_cm, weight_kg, weight_goal, activity_level, time_frame_months = get_user_inputs()

# Calculate BMI
bmi = weight_kg / ((height_cm / 100) ** 2)

# Calculate BMR (Basal Metabolic Rate) using Mifflin-St Jeor formula
if gender == "male":
    bmr = 10 * weight_kg + 6.25 * height_cm - 5 * age + 5
else:
    bmr = 10 * weight_kg + 6.25 * height_cm - 5 * age - 161

# Calculate TDEE (Total Daily Energy Expenditure) based on activity level
activity_multipliers = {
    "sedentary": 1.2,
    "light": 1.375,
    "moderate": 1.55,
    "active": 1.725,
    "very active": 1.9
}
tdee = bmr * activity_multipliers[activity_level]

# Calculate Ideal Body Weight (IBW) using Hamwi method
if gender == "male":
    ibw = 50 + (0.91 * (height_cm - 152.4))
else:
    ibw = 45.5 + (0.91 * (height_cm - 152.4))

# Calculate Daily Caloric Needs based on weight goal
if weight_goal == "loss":
    daily_caloric_needs = tdee - 500  # Deficit of 500 calories for weight loss
elif weight_goal == "gain":
    daily_caloric_needs = tdee + 500  # Surplus of 500 calories for weight gain
else:
    daily_caloric_needs = tdee

# Calculate macronutrient calories (protein, fat, carbohydrate)
protein_calories = (daily_caloric_needs * 0.2)
fat_calories = (daily_caloric_needs * 0.25)
carbohydrate_calories = (daily_caloric_needs * 0.55)


# Output the results
print("- Your BMI is {:.2f}".format(bmi))
print("- Your BMR is {:.2f} calories".format(bmr))
print("- Your TDEE is {:.2f} calories".format(tdee))
print("- Your Ideal Body Weight is {:.1f} kg".format(ibw))
print("- Your Daily Caloric Needs are {:.2f} calories".format(daily_caloric_needs))
print("- Protein Calories: {:.2f}".format(protein_calories))
print("- Fat Calories: {:.2f}".format(fat_calories))
print("- Carbohydrate Calories: {:.2f}".format(carbohydrate_calories))
```
